```python
from pathlib import Path
import random
import pandas as pd
import matplotlib.pyplot as plt
import torch
from PIL import Image

from model import ModelEspectreFase
from caracteristiques import obre_imatgeRGB, construeix_x0
from manifest import llista_imatges_recursiu

ARREL_DFFD = Path("/scratch/project_2018586/DFFD/ParentDir")
RUTA_MODEL = Path("/scratch/project_2018586/PhaseSpectrum/resultats/1VS1/stargan/model.pth")
RUTA_PREDICCIONS = Path("/scratch/project_2018586/PhaseSpectrum/resultats/1VS1/stargan/prediccions.csv")

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
MIDA_IMATGE = (192, 192)
LLINDAR = 0.5
LLAVOR = 42
```

```python
# carregam el model entrenat
model = ModelEspectreFase().to(DEVICE)
state_dict = torch.load(RUTA_MODEL, map_location=DEVICE)
model.load_state_dict(state_dict)
model.eval()
```

```python
@torch.no_grad()
def prediu_imatge(ruta_img):
    img_rgb = obre_imatgeRGB(ruta_img, mida=MIDA_IMATGE)
    x0, magnitud_log, fase_norm, _ = construeix_x0(img_rgb)

    x0 = x0.unsqueeze(0).to(DEVICE)
    magnitud_log = magnitud_log.unsqueeze(0).to(DEVICE)
    fase_norm = fase_norm.unsqueeze(0).to(DEVICE)

    logits = model(x0, magnitud_log, fase_norm)
    prob_fake = torch.sigmoid(logits).item()
    pred_fake = 1 if prob_fake >= LLINDAR else 0

    return prob_fake, pred_fake
```

```python
def selecciona_imatges_noves(n_real=5, n_fake=5, llavor=42):
    rng = random.Random(llavor)

    df_pred = pd.read_csv(RUTA_PREDICCIONS)
    rutes_ja_usades = set(df_pred["path"].astype(str))

    rutes_real = [
        p for p in llista_imatges_recursiu(ARREL_DFFD / "Real" / "ffhq")
        if str(p) not in rutes_ja_usades
    ]

    rutes_fake = [
        p for p in llista_imatges_recursiu(ARREL_DFFD / "Fake" / "stargan")
        if str(p) not in rutes_ja_usades
    ]

    seleccion_real = rng.sample(rutes_real, n_real)
    seleccion_fake = rng.sample(rutes_fake, n_fake)

    mostres = [(p, 0) for p in seleccion_real] + [(p, 1) for p in seleccion_fake]
    rng.shuffle(mostres)

    return mostres
```
